The Business Problem:

You're a quant analyst. Your boss doesn't want you to spend two weeks building an NLP model; they want you to spend two hours using a model that already works. They need you to score the sentiment of 10,000 news headlines today.

Load a pre-trained model called FinBERT. It has been trained by experts on millions of financial documents. It already understands the nuances of words like "bullish," "earnings," and "recession."

```python
# 1 - import the "easy button" (pipeline)
from transformers import pipeline
import pandas as pd

# 2 - Load the pre-trained FinBERT model
# The one line donwloads the model and st it up for the specifc task "Sentiment Analysis"
print("Loading FinBERT model..")
sentiment_pipeline = pipeline("sentiment-analysis", model="ProsusAI/finbert")
print("model loadded successfully")

# 3 - Testing it
# Create a list of financial headlines
headlines = [
    "Stock market hits all-time high as tech stocks soar.",
    "The company missed its earnings estimates, leading to a sell-off.",
    "Analysts remain neutral on the new product launch.",
    "FED announces unexpected interest rate hike, markets are fearful."
]

print("\nAnalyzing headlines..")
results = sentiment_pipeline(headlines)
print("Analysis complete")

# 4 - Print the results
df_results= pd.DataFrame(results)
print(df_results)
```

"Stock market hits all-time high...": Model classified this as positive (0.488), but it's very uncertain (almost 50/50).

"Analysts remain neutral...": Model classified this as negative (0.568).

Shows the nuance of these models. The specific model has learned to interpret "neutral" news as slightly "negative."
