The Project:

Financial Sentiment App


The Goal: To create a working web link that can click (or send to a friend) that opens a website. On that site, type a financial headline, and the AI tells if it's Positive, Negative, or Neutral.

```python
!pip install gradio
!pip install transformer

import gradio as gr
from transformers import pipeline
```

```python
# 1 - Load FinBERT
print("Loading FinBERT...")
classifier = pipeline("text-classification", model="ProsusAI/finbert")

# 2 - Define the handler
# Gradio function = input: String (text from user) --> Output: String (answer to user)

def analyze_financial_news(headline):
  # 2.1 - Pass text to model
  result = classifier(headline)[0]

  # 2.2 - Extract the label and score
  label = result['label']
  score = result['score']

  # 2.3 - Return a nice, formatted string
  return f"sentiment: {label.upper()} \nConfidence: {score:.2%}"

# 3 - Test - the function before building the UI
test_headline = "Stock market crashes as inflation rises."
print(f"\nTest Input: {test_headline}")
print(f"Test Output:\n{analyze_financial_news(test_headline)}")
```

The UI & Deployment

```python
# 1 - Build UI
demo = gr.Interface(
    fn=analyze_financial_news,
    inputs=gr.Textbox(label="Enter Financial Headline", placeholder="Type here.."),
    outputs=gr.Textbox(label="AI Sentiment Analysis"),
    title="💰 Financial Sentiment Analyzer",
    description="Enter a news headline to see if it's Bullish (Positive) or Bearish (Negative). Powered by FinBERT."
)

# 2 - Launch app
# share=True creates a public link (valid for 72 hours)
demo.launch(share=True)
```
